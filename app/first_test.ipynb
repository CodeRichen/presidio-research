```python
from presidio_evaluator.data_objects import InputSample, Span

# ====== 訓練資料（範例，實際訓練建議至少幾百筆才有效果） ======
train_samples = [
    InputSample(
        full_text="我的名字是老王，電話是0912345678",
        spans=[
            Span(entity_type="PERSON", entity_value="老王",
                 start_position=5, end_position=7),
            Span(entity_type="PHONE_NUMBER", entity_value="0912345678",
                 start_position=12, end_position=22),
        ],
        create_tags_from_span=True,
    ),
    InputSample(
        full_text="王小明住在台北市信義區",
        spans=[
            Span(entity_type="PERSON", entity_value="王小明",
                 start_position=0, end_position=3),
        ],
        create_tags_from_span=True,
    ),
    InputSample(
        full_text="請聯絡陳美麗，Email是test@example.com",
        spans=[
            Span(entity_type="PERSON", entity_value="陳美麗",
                 start_position=3, end_position=6),
            Span(entity_type="EMAIL_ADDRESS", entity_value="test@example.com",
                 start_position=10, end_position=27),
        ],
        create_tags_from_span=True,
    ),
    InputSample(
        full_text="今天天氣很好，適合出去玩",
        spans=[],  # 沒有 PII 的句子也要放一些，模型才學得會「不是每句都有PII」
        create_tags_from_span=True,
    ),
]

# ====== 驗證資料 ======
dev_samples = [
    InputSample(
        full_text="林志豪的電話是0987654321",
        spans=[
            Span(entity_type="PERSON", entity_value="林志豪",
                 start_position=0, end_position=3),
            Span(entity_type="PHONE_NUMBER", entity_value="0987654321",
                 start_position=8, end_position=18),
        ],
        create_tags_from_span=True,
    ),
    InputSample(
        full_text="這是一句沒有個資的普通句子",
        spans=[],
        create_tags_from_span=True,
    ),
]

print(f"train_samples: {len(train_samples)} 筆")
print(f"dev_samples: {len(dev_samples)} 筆")

# 檢查一下 tag 是否正確產生（確認 span position 沒抓錯字）
for s in train_samples:
    print(s.full_text)
    print(s.tags)
    print("---")
```

```python
from presidio_evaluator import InputSample

InputSample.create_spacy_dataset(train_samples, output_path="train.spacy")
InputSample.create_spacy_dataset(dev_samples, output_path="dev.spacy")
```
